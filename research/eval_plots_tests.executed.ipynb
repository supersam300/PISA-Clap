# eval_plots.py Smoke Tests

This notebook runs basic checks for the research plotting pipeline.

```python
from pathlib import Path
import subprocess
import sys

ROOT = Path('/Users/samarthatreya/Documents/Claped/PISA-Clap')
SCRIPT = ROOT / 'research' / 'eval_plots.py'
AUTO_CSV = ROOT / 'research' / 'abnormal_labels_auto.csv'

NORMAL_DIR = ROOT / 'model' / 'data' / 'normal'
ABNORMAL_DIR = ROOT / 'model' / 'data' / 'abnormal'

print('Python:', sys.executable)
print('Script exists:', SCRIPT.exists())
print('Normal dir exists:', NORMAL_DIR.exists())
print('Abnormal dir exists:', ABNORMAL_DIR.exists())
print('Auto CSV exists:', AUTO_CSV.exists())
```

```python
def run(cmd):
    print('\n$', ' '.join(cmd))
    out = subprocess.run(cmd, cwd=str(ROOT), capture_output=True, text=True)
    print('exit:', out.returncode)
    if out.stdout:
        print(out.stdout[-2000:])
    if out.stderr:
        print(out.stderr[-2000:])
    return out

r1 = run([sys.executable, '-m', 'py_compile', str(SCRIPT)])
assert r1.returncode == 0, 'py_compile failed'

r2 = run([sys.executable, str(SCRIPT), '--help'])
assert r2.returncode == 0, '--help failed'
assert 'Generate paper-ready CLAP/FFNN evaluation plots.' in (r2.stdout + r2.stderr)

print('\nSmoke tests passed.')
```

```python
# Optional small runtime check (may fail on local librosa/numba environment).
cmd = [
    sys.executable,
    str(SCRIPT),
    '--test-normal-dir', str(NORMAL_DIR),
    '--test-abnormal-dir', str(ABNORMAL_DIR),
    '--abnormal-label-csv', str(AUTO_CSV),
    '--max-samples', '1',
    '--latency-samples', '1',
    '--latency-repeats', '1',
]
r3 = run(cmd)
if r3.returncode != 0:
    print('Runtime check failed in current environment (expected on incompatible librosa/numba stack).')
else:
    print('Runtime check passed.')
```
